```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import datetime

import matplotlib as mpl
 
fig = plt.figure()
 
start = datetime.datetime(2016, 1, 1)
end = datetime.datetime(2017, 1, 1)
# 设置日期的间隔为1
delta = datetime.timedelta(days= 29)
 
# 生成一个matplotlib可以识别的日期对象
dates = mpl.dates.drange(start, end, delta)

print(dates)

date_format = mpl.dates.DateFormatter("%Y-%m")
print(date_format)
```

```python

# First figure
x1 = np.linspace(1, 12, 12)
x2 = np.tile(x1, 2)
x3 = np.linspace(1, 6, 6)
# x = np.tile(x, 2)
# itemNum = len(x)

x = np.concatenate((x2, x3), axis=0)
x = list(x)
x = ['2017-01','2017-02', '2017-03','2017-04','2017-05','2017-06','2017-07','2017-08','2017-09','2017-10','2017-11','2017-12',
    '2018-01', '2018-02', '2018-03', '2018-04', '2018-05', '2018-06', '2018-07', '2018-08', '2018-09', '2018-10', '2018-11', '2018-12',
    '2019-01', '2019-02', '2019-03', '2019-04', '2019-05', '2019-06', ]

plt.figure(figsize=(14,7))


# print(x)
itemNum = len(x)
print(itemNum)


# plt.subplot(2, 1, 1)
plt.plot(x, np.ones(itemNum), 'r-', label='tobacco related')
plt.plot(x, np.ones(itemNum)/2, 'g-', label='e-cigarette realated')
plt.plot(x, np.ones(itemNum)/3, 'b-', label='other tobacco')
plt.plot(x, np.ones(itemNum)/4, 'y-', label='first person')
plt.plot(x, np.ones(itemNum)/5, 'm-', label='not first person')
plt.legend()
# plt.vlines(6, 0, 1, colors = "black", linestyles = "dashed")
plt.axvline(5, hold=None,color='black',linestyle="--")
plt.annotate(s = 'Proposal', xy=(4,1))
plt.axvline(17, hold=None,color='black',linestyle="--")
plt.annotate(s = 'Approved', xy=(16,0.1))
plt.axvline(24, hold=None,color='black',linestyle="--")
plt.annotate(s = 'Enforcement ', xy=(23,0.1))




plt.xticks(rotation=80)
plt.xlabel('Month')
plt.ylabel('# of mentioning')
plt.title('Figure 1: Trends for number of tobacco related tweet by month. Y-axis represents the number of corresponding tweets within each month. ')
# plt.figure(figsize=(40,1))
plt.show()





# Second figure
# plt.subplot(2, 1, 2)
# plt.plot(x, np.sin(x), 'r-',
#         x, np.cos(x), 'g--')
# plt.xlabel('Month')
# plt.ylabel('# of mentioning')
# plt.title('Figure 2: Trends for proportion of tobacco related tweet by month. Y-axis represents proportion of all tweets.')
# plt.show()
```
