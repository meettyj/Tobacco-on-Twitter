```python
import numpy as np
def getAverageOnThreeMonths(data_list):
    if len(data_list) != 3:
        print('length should be 3')
        return -1
    return np.mean(data_list)

# def getAverageOnNineMonths(data_list):
#     if len(data_list) != 9:
#         print('length should be 3')
#         return -1
#     return np.mean(data_list)

#                                              state_ecig_in_all_ecig, state_ecig_in_all_state_tweets

# Based on three.
def calculateFrequencyProportion_on_threeMonths(allTweets, users_state, average_num_per_person_tweets_state, state_allECigarette, users_state_ecig,
                                            average_num_per_person_tweets_state_ecig,
                                            state_ecig_in_all_ecig, state_ecig_in_all_state_tweets):
    
    F_tweets = str(int(getAverageOnThreeMonths(allTweets[:3])))
    F_tweetPosters = str(int(getAverageOnThreeMonths(users_state[:3])))
    F_tweet_perPerson = str('%.3f' % getAverageOnThreeMonths(average_num_per_person_tweets_state[:3]))
    F_ecigTweets = str(int(getAverageOnThreeMonths(state_allECigarette[:3])))
    F_ecigTweetPosters = str(int(getAverageOnThreeMonths(users_state_ecig[:3])))
    F_ecigTweet_perPerson = str('%.3f' % getAverageOnThreeMonths(average_num_per_person_tweets_state_ecig[:3]))
    P_ecigTweets1 = str('%.3f' % getAverageOnThreeMonths(state_ecig_in_all_ecig[:3]))
    P_ecigTweets2 = str('%.5f' % getAverageOnThreeMonths(state_ecig_in_all_state_tweets[:3]))
    
    print('F_tweets: ', F_tweets)
    print('F_tweetPosters: ', F_tweetPosters)
    print('F_tweet_perPerson: ', F_tweet_perPerson)
    print('F_ecigTweets: ', F_ecigTweets)
    print('F_ecigTweetPosters: ', F_ecigTweetPosters)
    print('F_ecigTweet_perPerson: ', F_ecigTweet_perPerson)
    print('P_ecigTweets1: ', P_ecigTweets1)
    print('P_ecigTweets2: ', P_ecigTweets2)
    
    string_to_overleaf = ''+F_tweets+' & '+F_tweetPosters+' & '+F_tweet_perPerson+' & '+F_ecigTweets+' & '+F_ecigTweetPosters+' & '+F_ecigTweet_perPerson+' & '+P_ecigTweets1+' & '+P_ecigTweets2
    
    return string_to_overleaf

def getAverageOnSixMonths(data_list):
    if len(data_list) != 6:
        print('length should be 6')
        return -1
    return np.mean(data_list)


# Based on three months.
def calculateFrequencyProportion_on_sixMonths(allTweets, users_state, average_num_per_person_tweets_state, state_allECigarette, users_state_ecig,
                                            average_num_per_person_tweets_state_ecig,
                                            state_ecig_in_all_ecig, state_ecig_in_all_state_tweets):
    
    F_tweets = str(int(getAverageOnSixMonths(allTweets[:6])))
    F_tweetPosters = str(int(getAverageOnSixMonths(users_state[:6])))
    F_tweet_perPerson = str('%.3f' % getAverageOnSixMonths(average_num_per_person_tweets_state[:6]))
    F_ecigTweets = str(int(getAverageOnSixMonths(state_allECigarette[:6])))
    F_ecigTweetPosters = str(int(getAverageOnSixMonths(users_state_ecig[:6])))
    F_ecigTweet_perPerson = str('%.3f' % getAverageOnSixMonths(average_num_per_person_tweets_state_ecig[:6]))
    P_ecigTweets1 = str('%.3f' % getAverageOnSixMonths(state_ecig_in_all_ecig[:6]))
    P_ecigTweets2 = str('%.5f' % getAverageOnSixMonths(state_ecig_in_all_state_tweets[:6]))
    
    print('F_tweets: ', F_tweets)
    print('F_tweetPosters: ', F_tweetPosters)
    print('F_tweet_perPerson: ', F_tweet_perPerson)
    print('F_ecigTweets: ', F_ecigTweets)
    print('F_ecigTweetPosters: ', F_ecigTweetPosters)
    print('F_ecigTweet_perPerson: ', F_ecigTweet_perPerson)
    print('P_ecigTweets1: ', P_ecigTweets1)
    print('P_ecigTweets2: ', P_ecigTweets2)
    
    string_to_overleaf = ''+F_tweets+' & '+F_tweetPosters+' & '+F_tweet_perPerson+' & '+F_ecigTweets+' & '+F_ecigTweetPosters+' & '+F_ecigTweet_perPerson+' & '+P_ecigTweets1+' & '+P_ecigTweets2
    
    return string_to_overleaf
```
