```python
# Flavor keyword with some space included.
fruit_flavor_ecig = ['blueberry', 'pomegranate', 'strawberry', ' orange', 'candy,', 'grapefruit,', 'kiwi', 'guave,', 
                     'lemonade,', 'apple,', 'mango ', ' peach ', 'pineapple,', 'raspberry', 'mint ', 'lemon,', 
                     'grape,', 'watermelon', 'fruit', 'citrus', 'lime', 'banana', 'coconut', 'berry', 'dragon fruit'
                     'melon', 'cherry', 'menthol']
candy_and_desert_flavor_ecig = ['vanilla', 'gummy', 'cotton candy', 'butterscotch', 'candy cane', 'caramel', 
                                ' tart', 'cheesecake', 'cinnamon roll', 'macaron', 'cream', 'cookie', 'cake']
beverage_flavor_ecig = ['coffee', 'espresso', 'milk', 'cracker', 'mocha', 'cappuccino', 'cocoa']
others_flavor_ecig = ['flavor', 'pod'] # 'eon pod', 'cali pod', 'sea pod', 'ziip pod', 

# flavor_ecig.extend(fruit_flavor_ecig)
not_flavor_ecig = ['smoky', 'smoked']
```

```python
# State_name = 'Kansas'
monthList = ['01','02','03','04','05','06']
# State_name = 'NewJersey'
# monthList = ['04','05','06']
```

```python
def checkFlavorECigaretteInLine(line):
    try:
        lineInJson = json.loads(line)
    except:
        print('Cannot load the line: ', line)
        return False
    textInLine = lineInJson["text"]
#     print(textInLine)
    textAfterRE = re.sub('https?:\/\/t\.co\/[\s\S]{10}', '', textInLine)
    lowercaseLine = textAfterRE.lower()
#     print('----')
#     print(lowercaseLine)
#     print('*******************')
    
    # test for synonym
    if 'smok' not in lowercaseLine:
        return 0
    else:
        for not_flavor_keyword in not_flavor_ecig:
            if not_flavor_keyword in lowercaseLine:
                return 0
        # have smok but not smoky/smoked
        for fruit_flavor_keyword in fruit_flavor_ecig:
            if fruit_flavor_keyword in lowercaseLine:
                return 1
        for candy_and_desert_flavor_keyword in candy_and_desert_flavor_ecig:
            if candy_and_desert_flavor_keyword in lowercaseLine:
                return 2
        for beverage_flavor_keyword in beverage_flavor_ecig:
            if beverage_flavor_keyword in lowercaseLine:
                return 3
        for others_flavor_keyword in others_flavor_ecig:
            if others_flavor_keyword in lowercaseLine:
                return 4

    return 0
```

```python
def checkECigaretteOfEachLineInFile_version_2(fileName, month, State_name):


    
    fruit_flavor_list = []
    candy_and_desert_flavor_list = []
    beverage_flavor_list = []
    others_flavor_list = []

    allFlavor_list = []
    
    with open(fileName, "r") as f:
        everyLines = f.readlines()
        print('length of this file: ', len(everyLines))
        for line in everyLines:
           
            # Filter out those e-cig flavor
            status = checkFlavorECigaretteInLine(line)
            if status == 0:
                continue
            elif status == 1:
                fruit_flavor_list.append(line)
            elif status == 2:
                candy_and_desert_flavor_list.append(line)    
            elif status == 3:
                beverage_flavor_list.append(line)
            elif status == 4:
                others_flavor_list.append(line)
    
    # remove duplication
    fruit_flavor_set = set(fruit_flavor_list)
    candy_and_desert_flavor_set = set(candy_and_desert_flavor_list)
    beverage_flavor_set = set(beverage_flavor_list)
    others_flavor_set = set(others_flavor_list)
    
    # output the result
    print('length of fruit_flavor_set: ', len(fruit_flavor_set))
    print('length of candy_and_desert_flavor_set: ', len(candy_and_desert_flavor_set))
    print('length of beverage_flavor_set: ', len(beverage_flavor_set))
    print('length of others_flavor_set: ', len(others_flavor_set))
    
    # output the specific information
#     print('*******************************')
#     print('fruit_flavor_set: ')
#     for i in fruit_flavor_set:
#         lineInJson = json.loads(i)
#         textInLine = lineInJson["text"]
#         textAfterRE = re.sub('https?:\/\/t\.co\/[\s\S]{10}', '', textInLine)
#         print(textAfterRE)
#         print('---------------------')
        
#     print('*******************************')
#     print('candy_and_desert_flavor_set: ')
#     for i in candy_and_desert_flavor_set:
#         lineInJson = json.loads(i)
#         textInLine = lineInJson["text"]
#         textAfterRE = re.sub('https?:\/\/t\.co\/[\s\S]{10}', '', textInLine)
#         print(textAfterRE)
#         print('---------------------')
    
#     print('*******************************')
#     print('beverage_flavor_set: ')
#     for i in beverage_flavor_set:
#         lineInJson = json.loads(i)
#         textInLine = lineInJson["text"]
#         textAfterRE = re.sub('https?:\/\/t\.co\/[\s\S]{10}', '', textInLine)
#         print(textAfterRE)
#         print('---------------------')
    
#     print('*******************************')
    print('others_flavor_set: ')
    for i in others_flavor_set:
        lineInJson = json.loads(i)
        textInLine = lineInJson["text"]
        textAfterRE = re.sub('https?:\/\/t\.co\/[\s\S]{10}', '', textInLine)
        print(textAfterRE)
        print('---------------------')
        
#     # Write E-Cigarette data to local file
#     all_Flavor_ECigarette_FileName = '/mnt/volume/result/Flavor/juliana_all_Flavor_ECigarette_'+State_name+'_2017_'+str(month)+'.json'
#     # allLocation_FileName = './test_allECigarette_2019_01'
#     with open(all_Flavor_ECigarette_FileName, "w", encoding="utf-8") as f:
#         for line in allFlavorECigarette_Line:
#             f.write(str(line))
```
