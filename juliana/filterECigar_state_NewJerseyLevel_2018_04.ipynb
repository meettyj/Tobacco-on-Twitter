```python
def checkECigaretteInLine(line):
    try:
        lineInJson = json.loads(line)
    except:
        print('Cannot load the line: ', line)
        return False
    textInLine = lineInJson["text"]
    lowercaseLine = textInLine.lower()
    
    # test for synonym
    for keyword in synonym_ecig:
        #         print(keyword)
        if '&' not in keyword and keyword in lowercaseLine:
            return True
        elif '&' in keyword:
            k1, k2 = keyword.split('&')
            if k1 in lowercaseLine and k2 in lowercaseLine:
                return True

            # synonym_ecig_list.append(line)
            # flagHit = True
            # break

    # test for brand
    for keyword in brand_ecig:
        if '&' not in keyword and keyword in lowercaseLine:
            return True
        elif '&' in keyword:
            k1, k2 = keyword.split('&')
            if k1 in lowercaseLine and k2 in lowercaseLine:
                return True
#         if keyword in lowercaseLine:
#             return True
            # brand_ecig_list.append(line)
            # flagHit = True
            # break

    # test for policy
    for keyword in policy_ecig:
        if '&' not in keyword and keyword in lowercaseLine:
            return True
        elif '&' in keyword:
            k1, k2 = keyword.split('&')
            if k1 in lowercaseLine and k2 in lowercaseLine:
                return True
#         if keyword in lowercaseLine:
#             return True
            # policy_ecig_list.append(line)
            # flagHit = True
            # break

    # test for cessation
    for keyword in cessation_ecig:
        if '&' not in keyword and keyword in lowercaseLine:
            return True
        elif '&' in keyword:
            k1, k2 = keyword.split('&')
            if k1 in lowercaseLine and k2 in lowercaseLine:
                return True
#         if keyword in lowercaseLine:
#             return True
            # cessation_ecig_list.append(line)
            # flagHit = True
            # break

    return False
```

```python
def writeStateTweetsToDisk(State_name, allState, allECigarette_State_Line, allECigarette_not_State_Line, not_ECigarette_State_Line):
    State_name = str(State_name)
    
    # Write State data to local file
    # os.chdir('/mnt/volume/result/')
    allState_FileName = './juliana_all'+State_name+'_2018_'+month+'.json'
    with open(allState_FileName, "a", encoding="utf-8") as f:
        for line in allState:
            f.write(str(line))

    # Write E-Cigarette that happened in State data to local file
    allECigarette_State_FileName = './juliana_'+State_name+'_allECigarette_2018_'+month+'_categoryKeyWord.json'
    with open(allECigarette_State_FileName, "a", encoding="utf-8") as f:
        for line in allECigarette_State_Line:
            f.write(str(line))
    
    # Write E-Cigarette that not in State data to local file
    allECigarette_not_State_FileName = './juliana_not_'+State_name+'_allECigarette_2018_'+month+'_categoryKeyWord.json'
    with open(allECigarette_not_State_FileName, "a", encoding="utf-8") as f:
        for line in allECigarette_not_State_Line:
            f.write(str(line))
            
    # Write not E-cig in State to local file.
    not_ECigarette_State_FileName = './juliana_not_ECigarette_'+State_name+'_2018_'+month+'_categoryKeyWord.json'
    with open(not_ECigarette_State_FileName, "a", encoding="utf-8") as f:
        for line in not_ECigarette_State_Line:
            f.write(str(line))
```

```python
def checkECigaretteOfEachLineInFile(fileName):

    os.chdir('/mnt/volume/juliana/data')
    
    # New Jersey 
    allState_1 = set()
    allECigarette_State_1_Line = set()
    allECigarette_not_State_1_Line = set()
    not_ECigarette_State_1_Line = set()
    
    # New York
    allState_2 = set()
    allECigarette_State_2_Line = set()
    allECigarette_not_State_2_Line = set()
    not_ECigarette_State_2_Line = set()

    # Pennsylvania
    allState_3 = set()
    allECigarette_State_3_Line = set()
    allECigarette_not_State_3_Line = set()
    not_ECigarette_State_3_Line = set()
    
    # Connecticut
    allState_4 = set()
    allECigarette_State_4_Line = set()
    allECigarette_not_State_4_Line = set()
    not_ECigarette_State_4_Line = set()
    
    # Delaware
    allState_5 = set()
    allECigarette_State_5_Line = set()
    allECigarette_not_State_5_Line = set()
    not_ECigarette_State_5_Line = set()
    
    # Maryland
    allState_6 = set()
    allECigarette_State_6_Line = set()
    allECigarette_not_State_6_Line = set()
    not_ECigarette_State_6_Line = set()
    
    with open(fileName, "r") as f:
        everyLines = f.readlines()
        for line in everyLines:
            try:
                tweetInJson = json.loads(line)
                tweetState = getState(tweetInJson)
            except:
                continue          
            
            # First Level
            if checkECigaretteInLine(line):
                # State_1
                if tweetState == State_abbreviation_1:
                    allECigarette_State_1_Line.add(line)
                else:
                    allECigarette_not_State_1_Line.add(line)
                    
                # State_2
                if tweetState == State_abbreviation_2:
                    allECigarette_State_2_Line.add(line)
                else:
                    allECigarette_not_State_2_Line.add(line)
                    
                # State_3
                if tweetState == State_abbreviation_3:
                    allECigarette_State_3_Line.add(line)
                else:
                    allECigarette_not_State_3_Line.add(line)
                    
                # State_4
                if tweetState == State_abbreviation_4:
                    allECigarette_State_4_Line.add(line)
                else:
                    allECigarette_not_State_4_Line.add(line)

                # State_5
                if tweetState == State_abbreviation_5:
                    allECigarette_State_5_Line.add(line)
                else:
                    allECigarette_not_State_5_Line.add(line)     
                    
                # State_6
                if tweetState == State_abbreviation_6:
                    allECigarette_State_6_Line.add(line)
                else:
                    allECigarette_not_State_6_Line.add(line)  
                    
            # Second Level
            if tweetState == State_abbreviation_1:
                allState_1.add(line)
                if not checkECigaretteInLine(line):
                    not_ECigarette_State_1_Line.add(line)
            elif tweetState == State_abbreviation_2:
                allState_2.add(line)
                if not checkECigaretteInLine(line):
                    not_ECigarette_State_2_Line.add(line)
            elif tweetState == State_abbreviation_3:
                allState_3.add(line)
                if not checkECigaretteInLine(line):
                    not_ECigarette_State_3_Line.add(line)
            elif tweetState == State_abbreviation_4:
                allState_4.add(line)
                if not checkECigaretteInLine(line):
                    not_ECigarette_State_4_Line.add(line)
            elif tweetState == State_abbreviation_5:
                allState_5.add(line)
                if not checkECigaretteInLine(line):
                    not_ECigarette_State_5_Line.add(line)
            elif tweetState == State_abbreviation_6:
                allState_6.add(line)
                if not checkECigaretteInLine(line):
                    not_ECigarette_State_6_Line.add(line)
                    
    os.chdir('/mnt/volume/result/State_NewJerseyLevel/')


    writeStateTweetsToDisk(State_name_1, allState_1, allECigarette_State_1_Line, 
                           allECigarette_not_State_1_Line, not_ECigarette_State_1_Line)
    writeStateTweetsToDisk(State_name_2, allState_2, allECigarette_State_2_Line, 
                           allECigarette_not_State_2_Line, not_ECigarette_State_2_Line)
    writeStateTweetsToDisk(State_name_3, allState_3, allECigarette_State_3_Line, 
                           allECigarette_not_State_3_Line, not_ECigarette_State_3_Line)
    writeStateTweetsToDisk(State_name_4, allState_4, allECigarette_State_4_Line, 
                           allECigarette_not_State_4_Line, not_ECigarette_State_4_Line)
    writeStateTweetsToDisk(State_name_5, allState_5, allECigarette_State_5_Line, 
                           allECigarette_not_State_5_Line, not_ECigarette_State_5_Line)
    writeStateTweetsToDisk(State_name_6, allState_6, allECigarette_State_6_Line, 
                           allECigarette_not_State_6_Line, not_ECigarette_State_6_Line)    
    
            
```

```python
# Iterate the files in the directory you stored the data.
jsonFileDirectoryPath = "/mnt/volume/juliana/data"
dirNames = os.listdir(jsonFileDirectoryPath)

os.chdir('/mnt/volume/juliana/data')

for dirName in dirNames:
    if dirName.startswith('2018-'+month):
        #     print(dirName)
        files = os.listdir(dirName)
        for file in files:
#             print(file)
            file = dirName + '/' + file
            if not os.path.isdir(file) and (os.path.splitext(file)[-1] == ".json"):

                checkECigaretteOfEachLineInFile(file)
                print('finish file: ', file)
        os.chdir('/mnt/volume/juliana/data')

print('Finished all the files.')
```

```python
# Count lines of each file above.
def countLinesOfFile(State_name, month):
    State_name = str(State_name)
    length_of_juliana_allState = len(
        open('/mnt/volume/result/State_NewJerseyLevel/juliana_all'+State_name+'_2018_'+month+'.json', 'rU').readlines())
    length_of_juliana_State_allECigarette = len(open(
        '/mnt/volume/result/State_NewJerseyLevel/juliana_'+State_name+'_allECigarette_2018_'+month+'_categoryKeyWord.json', 'rU').readlines())
    length_of_juliana_not_State_allECigarette = len(open(
        '/mnt/volume/result/State_NewJerseyLevel/juliana_not_'+State_name+'_allECigarette_2018_'+month+'_categoryKeyWord.json', 'rU').readlines())
    length_of_not_ECigarette_State_FileName = len(open(
        '/mnt/volume/result/State_NewJerseyLevel/juliana_not_ECigarette_'+State_name+'_2018_'+month+'_categoryKeyWord.json', 'rU').readlines())
    
    print('month: ', month)
    print('juliana_all'+State_name+': ', length_of_juliana_allState)
    print('juliana_'+State_name+'_allECigarette: ', length_of_juliana_State_allECigarette)
    print('juliana_not_'+State_name+'_allECigarette: ', length_of_juliana_not_State_allECigarette)
    print('juliana_not_ECigarette_'+State_name+': ', length_of_not_ECigarette_State_FileName)
    
#     print('Writing to file in disk ...')
    # Write in seperate file.
    with open('/mnt/volume/result/State_NewJerseyLevel/result_2018_'+State_name, "a", encoding="utf-8") as f:
        f.write(str('month: ') + str(month) +'\n')
        f.write(str('juliana_all'+State_name+': ') + str(length_of_juliana_allState) +'\n')
        f.write(str('juliana_'+State_name+'_allECigarette: ') + str(length_of_juliana_State_allECigarette) +'\n')
        f.write(str('juliana_not_'+State_name+'_allECigarette: ') + str(length_of_juliana_not_State_allECigarette) +'\n')
        f.write(str('juliana_not_ECigarette_'+State_name+': ') + str(length_of_not_ECigarette_State_FileName) +'\n')
    
    print('Finished writing to disk.')
    
    
```

```python
# Count lines of each state files.
countLinesOfFile(State_name_1, month)
print('-------------------------')
countLinesOfFile(State_name_2, month)
print('-------------------------')
countLinesOfFile(State_name_3, month)
print('-------------------------')
countLinesOfFile(State_name_4, month)
print('-------------------------')
countLinesOfFile(State_name_5, month)
print('-------------------------')
countLinesOfFile(State_name_6, month)
print('-------------------------')
```
