```python
text_newKeyWord_list = []

# Another way of reading data.
with open('./data/juliana_SF_allECigarette_2019_01_categoryKeyWord.json', 'r', encoding="utf-8") as f:
    lines = f.readlines()
    for line in lines:
        lineInJson = json.loads(line)
        text_lineInJson = lineInJson["text"]
        text_newKeyWord_list.append(text_lineInJson)
#         text_newKeyWord_list.append(str(lineInJson))
        
        
        
#         lowercaseLine = line.lower()
```

```python
synonym_ecig_list = []
brand_ecig_list = []
policy_ecig_list = []
cessation_ecig_list = []
whatElse = []

# for line in new_text_df:
for line in text_newKeyWord_list:
    lowercaseLine = line.lower()
    
    flagHit = False
    
    # test for synonym
    count = 0
    for keyword in synonym_ecig:
#         count += 1
#         print(keyword)
        if keyword in lowercaseLine:
            synonym_ecig_list.append(line)
            flagHit = True
#             print(count)
            break # break the for loop.

    
    # test for brand
    for keyword in brand_ecig:
        if keyword in lowercaseLine:
            brand_ecig_list.append(line)
            flagHit = True
            break
        
    # test for policy
    for keyword in policy_ecig:
        if keyword in lowercaseLine:
            policy_ecig_list.append(line)
            flagHit = True
            break
        
    # test for cessation
    for keyword in cessation_ecig:
        if keyword in lowercaseLine:
            cessation_ecig_list.append(line)
            flagHit = True
            break
    
    if flagHit == False:
        whatElse.append(line)
    
print('length of synonym_ecig_list: ', len(synonym_ecig_list))
print('length of brand_ecig_list: ', len(brand_ecig_list))
print('length of policy_ecig_list: ', len(policy_ecig_list))
print('length of cessation_ecig_list: ', len(cessation_ecig_list))
            
```
