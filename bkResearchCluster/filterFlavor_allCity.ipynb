```python
# Didn't consider the conbination of ecig synonym and flavor, which may contain in the result of e-cig keyword filtering. 
def checkFlavorECigaretteInLine(line):
    try:
        lineInJson = json.loads(line)
    except:
        print('Cannot load the line: ', line)
        return False
    textInLine = lineInJson["text"]
    lowercaseLine = textInLine.lower()
    if 'smok' in lowercaseLine:
        for keyword in flavor_ecig:
            if keyword in lowercaseLine:
                return True
    return False
```

```python
def checkECigaretteOfEachLineInFile(fileName):

    os.chdir('/mnt/volume/juliana/data')
#     allECigaretteLine = set()
#     allSF = set()
#     allECigarette_SF_Line = set()
#     allECigarette_not_SF_Line = set()

    allFlavorECigarette_Line = set()
    
    with open(fileName, "r") as f:
        everyLines = f.readlines()
        for line in everyLines:
           
            # Filter out those e-cig flavor
            if checkFlavorECigaretteInLine(line):
                allFlavorECigarette_Line.add(line)
           
    os.chdir('/mnt/volume/result/')

    # Write E-Cigarette data to local file
    all_Flavor_ECigarette_FileName = './juliana_all_Flavor_ECigarette_2018_01.json'
    # allLocation_FileName = './test_allECigarette_2019_01'
    with open(all_Flavor_ECigarette_FileName, "a", encoding="utf-8") as f:
        for line in allFlavorECigarette_Line:
            f.write(str(line))
```

```python
# Iterate the files in the directory you stored the data.
jsonFileDirectoryPath = "/mnt/volume/juliana/data"
dirNames = os.listdir(jsonFileDirectoryPath)

os.chdir('/mnt/volume/juliana/data')

for dirName in dirNames:
    if dirName.startswith('2018-01'):
        #     print(dirName)
        files = os.listdir(dirName)
        for file in files:
#             print(file)
            file = dirName + '/' + file
            if not os.path.isdir(file) and (os.path.splitext(file)[-1] == ".json"):

                checkECigaretteOfEachLineInFile(file)
                print('finish file: ', file)
        os.chdir('/mnt/volume/juliana/data')

print('Finished all the files.')
```
